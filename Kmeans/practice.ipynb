```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris
```

```python
X = iris.drop(labels = 'Species',axis=1)
X
```

```python
kmeans = KMeans(n_clusters = 3 )
kmeans.fit(X)
X['cluster'] = kmeans.labels_
X.cluster.value_counts()
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

centers = kmeans.cluster_centers_
sns.lmplot(x = 'Petal_Length',y='Petal_Width',hue = 'cluster',markers=['^','s','o'],
          data = X ,fit_reg = False,scatter_kws = {'alpha':0.8},legend_out=False)

#legend_out是将图例存放在图框内
#scatter_kws是透明度
#fit_reg是是否要回归线
plt.scatter(centers[:,2],centers[:,3],marker='*',color='black',s=130)
plt.xlabel('花瓣长度')
plt.ylabel('花瓣宽度')
plt.grid()


plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

centers = kmeans.cluster_centers_
sns.lmplot(x = 'Sepal_Length',y='Sepal_Width',hue = 'cluster',markers=['^','s','o'],
          data = X ,fit_reg = False,scatter_kws = {'alpha':0.4},legend_out=False)

#legend_out是将图例存放在图框内
#scatter_kws是透明度
#fit_reg是是否要回归线
plt.scatter(centers[:,0],centers[:,1],marker='*',color='black',s=130)
plt.xlabel('花萼长度')
plt.ylabel('花萼宽度')
plt.grid()


plt.show()
```
